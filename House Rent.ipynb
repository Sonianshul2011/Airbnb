# Finding Missing Values

```python
missing_values= ["N/a","na",np.nan]
df = pd.read_csv("House_Rent_Dataset.csv" , na_values=missing_values)
```

```python
df.isnull().sum()
```

### In this Heat map, we can see that there is no missing values or null values i.e. there is only one colour in graph.

```python
sns.heatmap(df.isnull(), yticklabels=False)
```

### Drop na values.

```python
df.dropna()
```

### Fill the na values with the help of forward fill method.

```python
df.fillna(method='ffill')
```

### Fill the na values with the help of backward fill method.

```python
df.fillna(method='bfill')
```

### Interpolate the values

```python
df.interpolate()
```
